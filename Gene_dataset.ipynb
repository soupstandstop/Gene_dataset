```python
import requests as rq
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys 
import re
import time
```

## 抓Entrez相似的基因以及其別名，作為Dataset

```python
topic_gene = webdriver.Chrome("/Users/soup/PycharmProjects/TREC2018/chromedriver")
topic_gene.get('http://www.trec-cds.org/topics2017.xml')
gen = BeautifulSoup(topic_gene.page_source, "html.parser")
gene = gen.find_all("gene")

Topic1=[];Topic2=[];Topic3=[];Topic4=[];Topic5=[];Topic6=[];Topic7=[];Topic8=[];
Topic9=[];Topic10=[];Topic11=[];Topic12=[];Topic13=[];Topic14=[];Topic15=[];
Topic16=[];Topic17=[];Topic18=[];Topic19=[];Topic20=[];Topic21=[];Topic22=[];
Topic23=[];Topic24=[];Topic25=[];Topic26=[];Topic27=[];Topic28=[];Topic29=[];Topic30=[]
list_topic_gene = [Topic1,Topic2,Topic3,Topic4,Topic5,Topic6,Topic7,Topic8,
Topic9,Topic10,Topic11,Topic12,Topic13,Topic14,Topic15,
Topic16,Topic17,Topic18,Topic19,Topic20,Topic21,Topic22,
Topic23,Topic24,Topic25,Topic26,Topic27,Topic28,Topic29,Topic30]

l1 = []
for gene1 in gene:
    l1.append(gene1.text)
l2 = []
for gene2 in l1:
    l2.append(re.split(r'[(),]', gene2))
l3 = []
for gene3 in l2:
    l3.append(list(filter(None, gene3)))

browser = webdriver.Chrome("/Users/soup/PycharmProjects/TREC2018/chromedriver")
browser.get('https://www.ncbi.nlm.nih.gov/gene')

genenum = 0
for kkey in l3:
    for kkkey in kkey:
        frame = browser.find_element_by_xpath("//input[@id='term']")
        frame.clear()
        frame.send_keys(kkkey + Keys.RETURN)

        soup = BeautifulSoup(browser.page_source, "html.parser")
        page = soup.find("h3", class_="page")
        if page is not None:
            pagenum = page.text
        else:
            pagenum = '1'

        for pagenums in pagenum.split(): # 取此gene搜尋之頁數
            if pagenums.isdigit():
                nums = int(pagenums)
            else:
                pass
            
        if nums == 1: # page只有一頁
            soup = BeautifulSoup(browser.page_source, "html.parser")
            aliase = soup.find_all("td", class_="")
            if not aliase: # aliase is empty list
                name = soup.find('tbody')
                if name is not None: # name is not empty list
                    namee = name.find_all('a')
                    for names in namee:
                        list_topic_gene[genenum].append(names.text)
                else: # name is empty list
                    pass
            
            else: # aliase is not empty list
                cnt = 0
                for aliases in aliase:
                    cnt += 1
                    if cnt%3 == 0:
                        list_topic_gene[genenum].append(aliases.text)
                    else:
                        pass
                name = soup.find('tbody').find_all('a')
                for names in name:
                    list_topic_gene[genenum].append(names.text)
                #time.sleep(10)
            
        else:
            for i in range(1, nums+1):
                soup = BeautifulSoup(browser.page_source, "html.parser")
                aliase = soup.find_all("td", class_="")

                cnt = 0
                for aliases in aliase:
                    cnt += 1
                    if cnt%3 == 0:
                        list_topic_gene[genenum].append(aliases.text)
                    else:
                        pass

                name = soup.find('tbody').find_all('a')
                for names in name:
                    list_topic_gene[genenum].append(names.text)

                if  i == nums:
                    pass
                    #time.sleep(10)
                else:
                    browser.find_element_by_link_text("Next >").click()
                    #time.sleep(10)

    genenum += 1
browser.quit()
```

## dataset刪掉空字串
